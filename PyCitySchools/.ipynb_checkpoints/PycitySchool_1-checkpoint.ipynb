```python
merge_df_1=pd.merge(left=students_df, right=school_df, 
how='left', left_on='school',
right_on='name')
merge_df_1.head()
```

```python
school_df_2= merge_df_1[['Student ID','school','type','size','budget','reading_score','math_score']]
school_df_2.head()
```

```python
#School Summary

school_summary=school_df_2.set_index('school').groupby(['school'])
#school_df.head()
school_types = school_df.set_index('name')['type']
#school_types
school_total_stu = school_summary['Student ID'].count()
#school_total_stu
school_total_budget=school_df.set_index('name')['budget']
#school_total_budget
per_stud_budget=school_total_budget/school_total_stu
#per_stud_budget
avg_math_school=school_summary['math_score'].mean()
#avg_math_school
avg_read_school=school_summary['reading_score'].mean()
#avg_read_school
pass_math=school_df_2[school_df_2['math_score']>70].groupby(['school']).count()['Student ID']
perc_pass_math=(pass_math/school_total_stu)*100
#perc_pass_math
pass_read=school_df_2[school_df_2['reading_score']>70].groupby(['school']).count()['Student ID']
perc_pass_read=(pass_read/school_total_stu)*100
#perc_pass_read
overall_pass=(perc_pass_read+perc_pass_math)/2
#overall_pass
SchoolSummary=pd.DataFrame({"School Type": school_types,
                                   "Total Students": school_total_stu,
                                   "Total Budget": school_total_budget,
                                    "Per Student budget":per_stud_budget,
                                   "Average Math Score": avg_math_school,
                                   "Average Reading Score": avg_read_school,
                                   "% Passing Math":perc_pass_math,
                                   "% Passing Reading":perc_pass_read,
                                   "% Overall Passing Rate": overall_pass
                                   
})






SchoolSummary=SchoolSummary[["School Type",
                                   "Total Students",
                                   "Total Budget",
                                    "Per Student budget",
                                   "Average Math Score",
                                   "Average Reading Score",
                                   "% Passing Math",
                                   "% Passing Reading",
                                   "% Overall Passing Rate"]]
SchoolSummary=SchoolSummary.round(2)
SchoolSummary

SchoolSummary["% Passing Math"] = SchoolSummary["% Passing Math"].map("{0:,.2f}%".format)
SchoolSummary["% Passing Reading"] = SchoolSummary["% Passing Reading"].map("{0:,.2f}%".format)
SchoolSummary["% Overall Passing Rate"] = SchoolSummary["% Overall Passing Rate"].map("{0:,.2f}%".format)
SchoolSummary

```

```python
#Top Performing Schools (By Passing Rate)
top_5_schools = SchoolSummary.sort_values(["% Overall Passing Rate"], ascending=False)
top_5_schools.head()
```
